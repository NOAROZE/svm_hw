```python
import numpy as np
```

```python
def get_distance(p, w, c):
        numerator = w[0]*p[0] + w[1]*p[1]  + w[2]*p[2] + w[3]*p[3] + c
        denominator = np.sqrt(w[0]**2 + w[1]**2 + w[2]**2 + w[3]**2) 
        return numerator / denominator

points = np.array([
    [0, 0, 0, 0],   # P1
    [2, 0, 0, 0],   # P2
    [1, 1, 1, 1],   # P3
    [3, -1, 0, 2],  # P4
    [0, 2, 1, 1],   # P5
    [4, 1, -1, 0],  # P6
    [1, 0, 2, 1],   # P7
    [2, 1, 0, 3],   # P8
    [0, -1, 0, 3],  # P9
    [3, 2, 1, 0],   # P10
], dtype=float)

w = np.array([1, -2, 1, 0.5])
c = -3

distance = []
above = []
below = []

for p in points:
    dist = get_distance(p, w, c)
    if dist > 0:
        above.append(dist)
        print(f'This point: {p} is above on the surface')
    elif dist == 0:
        print(f'This point: {p} is on the surface')
    else:
        below.append(dist)
        print(f'This point: {p} is below on the surface')
    distance.append(float(get_distance(p, w, c)))

    
print(f'distance from each point to the surface: {distance}')
print(f'The closest point to the surface which is above the surface is: {min(above)}')
print(f'The closest point to the surface which is below the surface is: {max(below)}')
```
